```python
from numpy.random import multivariate_normal
```

```python
var1 = 1
var2 = 1
cov = 0.9
cov_matrix = [[var1, cov],
              [cov, var2]]
data = multivariate_normal([0, 0], cov_matrix, size=1000)
```

```python
plt.scatter(data[:, 0], data[:, 1])
```

```python
N = 1000
x = np.random.normal(size=N)
eps = np.random.normal(size=N)
y = x + 0.1  * eps
plt.scatter(x, y)
```
