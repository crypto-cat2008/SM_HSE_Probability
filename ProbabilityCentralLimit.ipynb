```python
# Modeling Cental Limit Theorem

sample_size = 10
number_of_samples = 10000
means = []
zscores = []
EX = 0.5        # this value needs to correspond to the distribution (normal)
VarX = 1. / 12  # this value needs to correspond to the distribution (normal)

for i in range(number_of_samples):
    sample = np.random.uniform(low=0, high=1, size=sample_size)
    mean = sum(sample) / len(sample)
    zscore = (mean -  EX) * np.sqrt(sample_size) /  np.sqrt(VarX)
    means.append(mean)
    zscores.append(zscore)
```

```python
plt.xlim(-4, 4)
plt.hist(zscores, density=True, bins=50);
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x))
```

```python
size = 1000
sigma_x = np.sqrt(1)
sigma_y = np.sqrt(1)

x = np.linspace(-3, 3, size)
y = np.linspace(-3, 3, size)

x, y = np.meshgrid(x, y)
z = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x**2/(2*sigma_x**2)
     + y**2/(2*sigma_y**2))))

plt.contourf(x, y, z, cmap='Blues')
plt.colorbar()
plt.show()
```
